```python
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.title('info')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
x=[5,8,10]
y=[12,6,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='line one',linewidth=10)
plt.plot(x2,y2,'c',label='line two',linewidth=5)
plt.title('epic info')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.grid(color='k')
plt.show()
```

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='example1')
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='example2')
plt.legend()
plt.show()
```
